```python
import sys
import requests
import inspect
#sys.path.append("../lib/scikit-multiflow/src/")


#for idx, path in enumerate(sys.path, 1):
#    print(f'{idx} - {path}')
#    print(f'\nrequests module location - {requests.__file__}')

from skmultiflow.trees.hoeffding_adaptive_tree import HoeffdingAdaptiveTreeClassifier

from skmultiflow.data.file_stream import FileStream
from skmultiflow.evaluation.evaluate_prequential import EvaluatePrequential
import numpy as np
import os
print (os.path.abspath(os.getcwd()))
from skmultiflow.trees.hoeffding_tree import HoeffdingTreeClassifier
print (inspect.getfile(HoeffdingTreeClassifier))  
```

```python

# Setup the File Stream
stream = FileStream("../data_output/export_dataframe_0v4.csv", -1, 1)
#stream = FileStream("../data_output/balanced.csv", -1, 1)

cfiers = [HoeffdingTreeClassifier(split_criterion="gini",nominal_attributes=[0,1],split_confidence=1e-10), 
          HoeffdingAdaptiveTreeClassifier(nominal_attributes=[0,1],split_confidence=1e-10)
         ]

evaluator = EvaluatePrequential(pretrain_size=10000, max_samples=1000000, batch_size=1000, n_wait=200, max_time=10000,
                                 output_file='summary_ht_balanced.txt', show_plot=False, 
                                metrics=['accuracy','true_vs_predicted'#,
                                         # 'precision',
                                         #'precision','recall','f1',
                                         #'running_time','model_size'
                                        ])

models = evaluator.evaluate(stream=stream, model=cfiers, model_names=['HT', 'HAT']);
#0.8499
```
